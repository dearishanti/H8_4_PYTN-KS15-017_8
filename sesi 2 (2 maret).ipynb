```python
while True:
    break
    print("hello")
```

```python
i = 0
while True:
    print("i awal", i)
    i += 1
    if i > 5:
        break
    # print("i akhir", i)
```

```python
n = 5
while n > 0:
    n -= 1
    if n % 2 == 0:
        print(n)
        break
print("Loop ended.")
```

```python
n = 5
while n > 0:
    print("n awal", n)
    n -= 1
    if n % 2 != 0:
        continue
    print("n genap", n)
print("Loop ended.")
```

```python
n = 5
while n > 0:
    n -= 1
    print(n)
    if n == 2:
        break
else:
    print('Loop done.')
```

```python
n = 5
while n > 0:
    m = "foobar"
    n -= 1
    print(n)
    while n < len(m):
        print(m[n])
        break
    if True:
        print("x")
else:
    print('Loop done.')
```
